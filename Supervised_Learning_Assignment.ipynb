Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose.
- Simple Linear Regression (SLR) is a statistical method that models the relationship between one independent (predictor) variable and one dependent (outcome) variable using a straight line.
Simple Linear Regression helps to predict the value of one variable (Y) based on the value of another variable (X) **using a straight-line equation**:
Y = a + bX + e

Where:

Y = Dependent variable (the outcome or variable we want to predict)

X = Independent variable (the predictor or explanatory variable)

a = Intercept (value of Y when X = 0)

b = Slope (shows how much Y changes when X increases by 1 unit)

e = Error term (difference between the observed and predicted values)

 **Purpose of Simple Linear Regression**:

1. Prediction:
To predict the value of a dependent variable based on the known value of the independent variable.
Example: Predicting a person‚Äôs weight based on their height.


2. Understanding Relationships:
To determine whether and how strongly two variables are related.
Example: Analyzing how advertising expenditure (X) affects sales revenue (Y).


3. Trend Analysis:
To identify trends or directions in data over time (e.g., predicting future demand based on past sales).


4. Decision Making:
Helps businesses and researchers make informed decisions based on quantitative evidence.


5. Quantifying Effect:
The slope (b) tells how much the dependent variable changes with each unit change in the independent variable.







Question 2: What are the key assumptions of Simple Linear Regression?
- Key Assumptions of Simple Linear Regression (SLR)
1. **Linearity**:
The relationship between the independent variable (X) and the dependent variable (Y) should be linear.
This means that a change in X leads to a proportional change in Y.

Check: Use a scatter plot ‚Äî the points should roughly form a straight line.

2. **Independence of Errors**:

The residuals (errors) should be independent of each other.
This means that the error for one observation should not influence the error for another.

 Check: Use the Durbin‚ÄìWatson test (especially for time-series data).

3. **Homoscedasticity (Constant Variance of Errors)**:
The variance of residuals should be the same across all values of X.
In other words, the spread of errors should not increase or decrease with X.

 Check: Plot residuals vs. predicted values ‚Äî the spread should appear random and uniform.

 4. **Normality of Errors**:
The residuals (differences between observed and predicted values) should be normally distributed.
This is important for hypothesis testing and confidence intervals.

 Check: Use a histogram or Q‚ÄìQ plot of residuals.

5. **No or Minimal Multicollinearity**:

Although this mainly applies to multiple regression, in SLR it means the independent variable should not be a constant or perfectly correlated with another predictor (since there‚Äôs only one X).

 6. **Measurement Accuracy**:
Both the dependent and independent variables should be measured accurately without significant error.

Question 6: What is Logistic Regression? How does it differ from Linear Regression?
- **Logistic Regression** is a statistical method used to model the relationship between a categorical dependent variable (usually binary) and one or more independent variables.
Logistic regression differs from linear regression in its purpose and the type of dependent variable it predicts: linear regression is for predicting continuous outcomes (like house prices) using a linear relationship between variables, whereas logistic regression is for predicting categorical outcomes (like yes/no or spam/not-spam) by modeling the probability of an event occurring.

**Linear Regression**

- Purpose: Used for regression problems where the goal is to predict a continuous outcome.

- Dependent Variable: A continuous variable (e.g., price, temperature, weight).
Output: A continuous numerical value.

- Methodology: Calculates a straight line to represent the relationship between variables, often using ordinary least squares (OLS) for parameter estimation.

- Example: Predicting the sales revenue for a store based on advertising spend.

**Logistic Regression**

- Purpose: Used for classification problems, predicting a categorical or binary outcome.

- Dependent Variable: A categorical variable, most commonly binary (e.g., yes/no, true/false).

- Output: The probability of an event occurring, mapped to a value between 0 and 1. This probability is then often used with a threshold to assign a final category.

- Methodology: Uses the sigmoid (S-shaped) activation function to squash the output of a linear equation into a probability, employing maximum likelihood estimation for parameter estimation.

- Example: Predicting whether a customer will churn (yes/no) based on their service usage.


Question 7:Name and briefly describe three common evaluation metrics for regression¬†models.
- The Three common evaluation metrics used to assess the performance of regression models:
1. **Mean Absolute Error (MAE)**

image.png

Meaning: Average of the absolute differences between actual (ùëåùëñ) and predicted (ùëåùëñ^) values.

Interpretation:

- Measures how far predictions are, on average, from the actual values.

- Lower MAE ‚Üí better model performance.

Advantage: Easy to interpret; gives equal weight to all errors.

2. **Mean Squared Error (MSE)**

image.png

Meaning: Average of the squared differences between actual and predicted values.

Interpretation:

- Penalizes larger errors more heavily because of squaring.

= Lower MSE ‚Üí better fit.

Disadvantage: The units are squared, making it less interpretable directly.

3. **R-squared (ùëÖ2) ‚Äì Coefficient of Determination**

image.png

Meaning: Proportion of variance in the dependent variable that is explained by the model.

Interpretation:

- R2 =1: Perfect fit
- R2 = 0: Model explains none of the variability
- Higher ùëÖ2 ‚Üí better explanatory power

Advantage: Provides an intuitive measure of how well the model explains the data.
